```
import numpy as np
import pandas as pd

#For this project what I'm trying to do is to clean and transform the data in this dataset by using libraries like numpy and pandas, 
#The followings blocks I'll make some questions to face the problematics in this data set

#Use pandas to read the csv
df = pd.read_csv('/work/data/netflix1.csv')

df
```

```
#Is there any null values?
print(df.isnull().sum())

print('-----------------------')

print('If there any null value?', '\n', 'Answer: ',df.isnull().values.any(), '- Although there is a value that we can count as null known as "Not Given"')
```

```
#Use drop_duplicates method to remove all the duplicates in the dataframe
duplicated = df.drop_duplicates(subset='show_id')

print('Is there any duplicated value?')
print('----------------------------------------------------------------')
print('Dataframe before searching for duplicates: ',df['show_id'].count())
print('Dataframe after searching for duplicates: ',duplicated['show_id'].count())
print('----------------------------------------------------------------')
print('As you can see, there are not duplicates in this dataset')
```

```
print('In the column "listen_in" we have more than one category for some titles, is there a way to remove the other categories and just keep the first one?')
print('Indeed there is a way, by making a new column using the "listed_in" column first category')

#Make a new column and use str.split method to split the information and just take the first part of the split by using the str function positionating it at 0 as first argument
df['category'] = df['listed_in'].str.split(',').str[0]

df
```

```
print("Is there any column that doesn't contribute to give information to the analysis?")
print('Yes, we will remove the columns "show_id", "title" and "listed_in" since they are not useful for the analysis and visualization')

df = df.drop(columns=['show_id','title','listed_in'])
df

```
